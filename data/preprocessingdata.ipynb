```python
#jpg변환
for i in range(len(ids)):
    img_id = ids[i]
    try:
        if img_id[-4:]!=".jpg":
            im = Image.open(DATA_ROOT+'images/'+img_id).convert('RGB')
            im.save(DATA_ROOT+'images/'+img_id[:-4]+'.jpg', 'jpeg')
            os.remove(DATA_ROOT+'images/'+img_id)
    except:
        print(img_id)
        
 
        
        
    
```

```python
# xml없는 이미지 폴더에 파일이 있으면 error로 옮김
for i in range(len(ids)):
    img_id = ids[i]
    if os.path.isfile(annopath+img_id[:-4]+'.xml') == False:
        shutil.move(DATA_ROOT+'images/'+img_id, DATA_ROOT+'error1')
        print(img_id)
```

```python
#xml폴더 리스트
ids2=os.listdir(DATA_ROOT+'annotations')
```

```python
len(ids2)
```

```python
# 이미지없는 xml을 error2로 옮김
for i in range(len(ids)):
    xml_id = ids2[i]     
    if os.path.isfile(DATA_ROOT+'images/'+xml_id[:-4]+'.jpg') == False:
        shutil.move(annopath+xml_id, DATA_ROOT+'/error2')        
        print("clear2",xml_id)
```

```python
VOC_CLASSES = (  # always index 0
    'fire',
'smoke',
'person',
'firefighter'
)
```

```python
# 옮지 않는 class가 있는 경우 error3로 옮김
for i in range(len(ids2)):
    print("going : ", annopath+ids2[i])
    target = ET.parse(annopath+ids2[i]).getroot()
    for obj in target.iter('object'):
        name = obj.find('name').text.lower().strip()
        if name not in VOC_CLASSES:
            print("moving : " , annopath+ids2[i])
            print("name : " , name)
            shutil.move(annopath+ids2[i], DATA_ROOT+'/error3')        

```
